### <span style='color:blue'>11] CHECKING FOR NOT AVAILABLE VALUES AND X VALUES</span>

```python
columns_with_x = []
for col in data.columns:
    if data[col].dtype == object:
        x_count = data[col].str.count('X').sum()  # Count the number of 'X' values
        if x_count > 0:
            columns_with_x.append(col)
            print(f"Column '{col}' contains {x_count} occurrences of 'X'")
```

#### <font color=red>Inference:</font>
1. FirstTimeHomebuyer,PPM,PropertyType categorical columns contains X value. For better analysis we will replace those values with mode.
2. MSA,NumBorrowers,PostalCode numerical columns also contain X value that why their datatype mentioned as object.For better analysis we will replace those values with median

```python
for cat in ['FirstTimeHomebuyer','PPM','PropertyType']:
    data[cat]=data[cat].str.strip()
    data[cat]=data[cat].replace('X',data[cat].mode()[0])
```

```python
data['MSA'] = pd.to_numeric(data['MSA'], errors='coerce').fillna(0).astype(np.int64)
data['MSA'].replace(0,data['MSA'].median(),inplace=True) 
```

```python
data['NumBorrowers'] = pd.to_numeric(data['NumBorrowers'], errors='coerce').fillna(0).astype(np.int64)
data['NumBorrowers'].replace(0,data['NumBorrowers'].median(),inplace=True) 
```

### <span style='color:blue'>12]  CONVERT DATE COLUMNS INTO DATE AND TIME FORMAT </span>

```python
# Convert the date columns to datetime format
data['FirstPaymentDate'] = pd.to_datetime(data['FirstPaymentDate'],format='%Y%m')
data['MaturityDate'] = pd.to_datetime(data['MaturityDate'], format='%Y%m')
```

```python
data['FirstPaymentYear']=data['FirstPaymentDate'].dt.year
data['FirstPaymentMonth']=data['FirstPaymentDate'].dt.month   
data['MaturityYear']=data['MaturityDate'].dt.year
data['MaturityMonth']=data['MaturityDate'].dt.month
```

### <span style='color:blue'>13]  CREATING NEW COLUMNS</span>

```python
##Transforming  Numeric feature LTV to diff range

# Create a new column to store the classifications
data['LTV_Range'] = ""

# Iterate over the values in the numeric column and assign classifications
for index, value in data['LTV'].iteritems():
    if value >= 0 and value < 25:
        data.at[index, 'LTV_Range'] = 'Low'
    elif value >= 25 and value < 50:
        data.at[index, 'LTV_Range'] = 'Medium'
    elif value >= 50 and value < 1000:
        data.at[index, 'LTV_Range'] = 'High'
    
    else:
        data.at[index, 'LTV_Range'] = 'Invalid value'
```

```python
##Transforming  Numeric feature CreditScore to diff range

# Create a new column to store the classifications
data['CreditRange'] = ""

# Iterate over the values in the numeric column and assign classifications
for index, value in data['CreditScore'].iteritems():
    if value >= 0 and value < 650:
        data.at[index, 'CreditRange'] = 'Poor'
    elif value >= 650 and value < 700:
        data.at[index, 'CreditRange'] = 'Fair'
    elif value >= 700 and value < 750:
        data.at[index, 'CreditRange'] = 'Good'
    elif value >= 750 and value <= 900:
        data.at[index, 'CreditRange'] = 'Excellent'
    else:
        data.at[index, 'CreditRange'] = 'Invalid value'
```

```python
##Transforming  Numeric feature MonthsInRepayment to diff range

# Create a new column to store the classifications
data['RePayRange'] = ""

# Iterate over the values in the numeric column and assign classifications
for index, value in data['MonthsInRepayment'].iteritems():
    if value >= 0 and value < 48:
        data.at[index, 'RePayRange'] = '0-4yrs'
    elif value >= 48 and value < 96:
        data.at[index, 'RePayRange'] = '4-8yrs'
    elif value >= 96 and value < 144:
        data.at[index, 'RePayRange'] = '8-12yrs'
    elif value >= 144 and value < 192:
        data.at[index, 'RePayRange'] = '12-16yrs'
    elif value >= 192 and value < 240:
        data.at[index, 'RePayRange'] = '16-20yrs'
    else:
        data.at[index, 'RePayRange'] = 'Invalid value'
```

```python
# Dropping unnecessary column
data.drop(['PostalCode','LoanSeqNum','ProductType','FirstPaymentDate','MaturityDate','FirstTimeHomebuyer','LTV','CreditScore'],axis=1, inplace=True)
```
